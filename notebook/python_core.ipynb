# 1. <a id='toc1_'></a>[Mutable Default Argument](#toc0_)

Problem: the same list is reused across calls

```python
def add(x, lst=[]):
    lst.append(x)
    return lst

# Change this default value to "None" and initialize this parameter inside the function/method.sonarqube(ipython:S5717)
```

```python
t = add(1)
t
```

```python
t = add(2)
t
```

```python
# Solution
def add(x, lst=None):
    if lst is None:
        lst = []
    lst.append(x)
    return lst
```

```python
t = add(1)
t
```

```python
t = add(2)
t
```

# 2. <a id='toc2_'></a>[Function modifies input](#toc0_)

```python
def sort_data(data):
    data.sort()
```

```python
arr = [3,4,1]
sort_data(arr)
print(arr)
```

```python
# Solution
def sort_data(data):
    new_data = data[:]
    new_data.sort()
    return new_data
```

```python
arr = [3,4,1]
sorted_data = sort_data(arr)
print(arr)
print(sorted_data)
```

## 2.1. <a id='toc2_1_'></a>[Reassigning a variable is safe; mutating the object is not.](#toc0_)

🎯 The key difference

| Code             | What happens           | Affect outside? |
| ---------------- | ---------------------- | --------------- |
| `data = [2, 6]`  | New object             | ❌ No            |
| `data.append(6)` | Modify existing object | ✅ Yes           |

🧠 Mental model

👉 Variables are just **labels**

* `data = [...]` → move label to a new box
* `.append()` → change what’s inside the box

```python
def show_statistics(data):
    data = [2, 6]
    print(f"Output data: {data}")

a = [2,3]
print(f"step 1: a={a}")
show_statistics(a)
print(f"step 2: a={a}")
```

# 5. <a id='toc5_'></a>[Python decorators](#toc0_)

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} took  {time.time()-start_time}(s).")
        return result
    return wrapper
```

```python
@timer
def demo_function():
    time.sleep(2)
    return "Finished-Demo decorator"

demo_function()
```

# 7. <a id='toc7_'></a>[Explain the difference between repr and str](#toc0_)

🔹 What is the difference between `str` and `repr`?

Both are used to **convert an object into a string**, but they are meant for **different purposes**.

🔹 Easy way to understand

- `str()` → **for humans (readable, friendly)**
- `repr()` → **for developers (detailed, unambiguous)**

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old."

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"
```

```python
person = Person('LaVie', 32)
print(person)
```

```python
print(repr(person))
```

# 9. <a id='toc9_'></a>[How would you safely handle missing keys in a dictionary?](#toc0_)

In practice, the best approach depends on the situation:
+ get() is most concise when you need a simple default
+ in checks are clear when you need different logic based on presence
+ try/except follows Python’s “easier to ask forgiveness than permission” philosophy — meaning you just try the action and handle the error if it happens, instead of checking first
+ defaultdict is best when you’re building a collection

Note: For most simple cases, get() provides the cleanest solution.

```python
# Using the get() method with a default value
user_settings = {"theme": "dark"}
font_size = user_settings.get("font_size", 12)
font_size
```

```python
# Using try/except to catch KeyError
try:
    font_size = user_settings["font_size"]
except KeyError:
    font_size = 12

font_size
```

```python
# Using the in operator to check first
if "font_size" in user_settings:
    font_size = user_settings["font_size"]
else:
    font_size = 12

font_size
```

```python
# Using collections.defaultdict to provide automatic defaults
from collections import defaultdict

word_counts = defaultdict(int)
word_counts["hello"] += 1  # No KeyError even though "hello" wasn't there

word_counts
```

# 10. <a id='toc10_'></a>[What is the Global Interpreter Lock (GIL) and why does it matter?](#toc0_)

```python
# CPU-bound
import threading
import time

def cpu_task():
    total = 0
    for i in range(10_000_000):
        total += i

# Sequential
start_time = time.time()
cpu_task()
cpu_task()
print(f"Sequential done in {time.time() - start_time}(s)")


# Multi-threading
start_time = time.time()
t1 = threading.Thread(target=cpu_task)
t2 = threading.Thread(target=cpu_task)
t1.start()
t2.start()

t1.join()
t2.join()
print(f"Multithread done in {time.time() - start_time}(s)")
```

👉 **Expected outcome**:
The threaded version is **not significantly faster** than sequential execution.

**Why?**
Because both threads compete for the GIL, so only one runs Python bytecode at a time.

```python
# I/O-bound case
import threading
import time

def io_task():
    time.sleep(1)  # Simulate I/O wait

start_time = time.time()
t1 = threading.Thread(target=io_task)
t2 = threading.Thread(target=io_task)
t1.start()
t2.start()
t1.join()
t2.join()
print(f"Finished - I/O threaded: {time.time() - start_time}(s)")
```

👉 This runs in ~1 second (not 2), because:

- while one thread is waiting (sleep / I/O),
- the GIL is released → another thread can run

🧠 Key takeaway (say this explicitly)

> “Threads in Python are good for I/O concurrency, not CPU parallelism.”

🚀 What to do instead

- CPU-bound:
  - multiprocessing (multiple processes, multiple GILs)
  - native extensions (NumPy, C/C++)

- I/O-bound:
  - threading or asyncio (FastAPI, async frameworks)
