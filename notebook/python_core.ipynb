# 1. <a id='toc1_'></a>[Mutable Default Argument](#toc0_)

Problem: the same list is reused across calls

```python
#
def add(x, lst=[]):
    lst.append(x)
    return lst

# Change this default value to "None" and initialize this parameter inside the function/method.sonarqube(ipython:S5717)
```

```python
t = add(1)
t
```

```python
t = add(2)
t
```

```python
# Solution
def add(x, lst=None):
    if lst is None:
        lst = []
    lst.append(x)
    return lst
```

```python
t = add(1)
t
```

```python
t = add(2)
t
```

# 2. <a id='toc2_'></a>[Function modifies input](#toc0_)

```python
def sort_data(data):
    data.sort()
```

```python
arr = [3,4,1]
sort_data(arr)
print(arr)
```

```python
# Solution
def sort_data(data):
    new_data = data[:]
    new_data.sort()
    return new_data
```

```python
arr = [3,4,1]
sorted_data = sort_data(arr)
print(arr)
print(sorted_data)
```

## 2.1. <a id='toc2_1_'></a>[Reassigning a variable is safe; mutating the object is not.](#toc0_)

🎯 The key difference

| Code             | What happens           | Affect outside? |
| ---------------- | ---------------------- | --------------- |
| `data = [2, 6]`  | New object             | ❌ No            |
| `data.append(6)` | Modify existing object | ✅ Yes           |

🧠 Mental model

👉 Variables are just **labels**

* `data = [...]` → move label to a new box
* `.append()` → change what’s inside the box

```python
def show_statistics(data):
    data = [2, 6]
    print(f"Output data: {data}")

a = [2,3]
print(f"step 1: a={a}")
show_statistics(a)
print(f"step 2: a={a}")
```

# 3. <a id='toc3_'></a>[Shared reference](#toc0_)

```python
a = [1,3,2]
b = a
b[0] = -5
```

```python
a
```

```python
b
```

```python
# Solution
a = [1,3,2]
b = a[:]  # Copy array
b[0] = -5
print(f"a={a}")
print(f"b={b}")
```

# 4. <a id='toc4_'></a>[Hidden mutation (nested)](#toc0_)

👉 You didn’t change the variable — you changed what’s inside it

```python
def add_user(data):
    data["users"].append("A")

d = {"users": []}
add_user(d)
d
```

# 5. <a id='toc5_'></a>[Python decorators](#toc0_)

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} took  {time.time()-start_time}(s).")
        return result
    return wrapper
```

```python
@timer
def demo_function():
    time.sleep(2)
    return "Finished-Demo decorator"

demo_function()
```
